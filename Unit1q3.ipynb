```python
import sympy as sp

t = sp.symbols('t')
v = sp.Function('v')

# Bernoulli equation dv/dt + v = v^2  (n=2)
ode = sp.Eq(v(t).diff(t) + v(t), v(t)**2)
sol = sp.dsolve(ode, v(t), ics={v(0): sp.Rational(1, 2)})
print("v(t) =", sp.simplify(sol.rhs))

# Verify via substitution w = 1/v -> dw/dt - w = -1 (linear ODE)
w = sp.Function('w')
ode_w = sp.Eq(w(t).diff(t) - w(t), -1)
sol_w = sp.dsolve(ode_w, w(t), ics={w(0): 2})  # w(0) = 1/v(0) = 2
print("w(t) = 1/v(t) =", sp.simplify(sol_w.rhs))
print("=> v(t) =", sp.simplify(1/sol_w.rhs))
```
