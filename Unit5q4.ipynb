```python
import numpy as np
import matplotlib.pyplot as plt

V0, a, R, C = 5, 1, 1, 1

omega = np.linspace(-20, 20, 2000)
omega_safe = np.where(omega == 0, 1e-12, omega)

V_in = 2*V0*np.sin(omega_safe*a)/omega_safe
H = 1/(1 + 1j*omega*R*C)
V_out = H * V_in

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(omega, V_in)
plt.title('Input Spectrum V_in(omega)'); plt.xlabel('omega'); plt.grid(True)

plt.subplot(1,2,2)
plt.plot(omega, np.abs(V_out))
plt.title('Output Spectrum |V_out(omega)|'); plt.xlabel('omega'); plt.grid(True)

plt.tight_layout()
plt.savefig('p24_fourier_transform.png', dpi=120)
plt.show()











```
