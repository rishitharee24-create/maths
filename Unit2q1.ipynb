```python
import sympy as sp

t = sp.symbols('t', positive=True)
i = sp.Function('i')
R, L, C = 4, 1, sp.Rational(1, 4)

ode = sp.Eq(L*i(t).diff(t, 2) + R*i(t).diff(t) + i(t)/C, 0)
sol = sp.dsolve(ode, i(t), ics={i(0): 1, i(t).diff(t).subs(t, 0): 0})
print("i(t) =", sp.simplify(sol.rhs))

# Characteristic equation analysis
m = sp.symbols('m')
char_eq = L*m**2 + R*m + 1/C
roots = sp.solve(char_eq, m)
print("Roots:", roots)
disc = R**2 - 4*L/C
if disc > 0:
    print("Overdamped (two distinct real roots)")
elif disc == 0:
    print("Critically damped (repeated real root)")
else:
    print("Underdamped (complex conjugate roots)")


```
