```python
import numpy as np
import sympy as sp

A = np.array([
    [-3, -2],
    [2, 0]
])

eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues (modes):", eigvals)
print("Eigenvectors (mode shapes):\n", eigvecs)

# Symbolic general solution x(t) = c1*v1*e^(lambda1 t) + c2*v2*e^(lambda2 t)
t, c1, c2 = sp.symbols('t c1 c2')
lam1, lam2 = sp.Matrix(eigvals).applyfunc(sp.nsimplify)
print("\nGeneral free response:")
print("x(t) = c1 * v1 * exp(lambda1*t) + c2 * v2 * exp(lambda2*t)")
print(f"lambda1 = {eigvals[0]:.4f}, lambda2 = {eigvals[1]:.4f}")









```
