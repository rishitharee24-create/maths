```python
import sympy as sp

t, s = sp.symbols('t s', positive=True)
R, L, C = 3, 1, sp.Rational(1, 2)

H_s = 1 / (L*C*s**2 + R*C*s + 1)
H_s = sp.simplify(H_s)
print("H(s) =", H_s)

# Unit step input: V_in(s) = 1/s
V_out_s = sp.simplify(H_s / s)
print("V_out(s) =", V_out_s)

v_out_t = sp.simplify(sp.inverse_laplace_transform(V_out_s, s, t))
print("v_out(t) =", v_out_t)

# Poles -> damping behavior
poles = sp.solve(L*C*s**2 + R*C*s + 1, s)
print("Poles of H(s):", poles)





```
