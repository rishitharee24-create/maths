```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

t, n = sp.symbols('t n', positive=True)
L = sp.pi
f = t  # sawtooth on (0, L)

bn = (2/L) * sp.integrate(f * sp.sin(n*sp.pi*t/L), (t, 0, L))
bn_simplified = sp.simplify(bn)
print("b_n =", bn_simplified)

# Evaluate first 5 coefficients
bn_func = sp.lambdify(n, bn_simplified, 'numpy')
N = 5
coeffs = [bn_func(k) for k in range(1, N+1)]
print("First 5 coefficients:", coeffs)

# Plot reconstruction
tvals = np.linspace(0, np.pi, 300)
v_approx = np.zeros_like(tvals)
for k in range(1, N+1):
    v_approx += coeffs[k-1] * np.sin(k*np.pi*tvals/np.pi)

plt.plot(tvals, tvals, '--', label='Original v(t)=t')
plt.plot(tvals, v_approx, label=f'{N}-term sine series')
plt.legend(); plt.grid(True); plt.xlabel('t'); plt.title('Half-Range Sine Series')
plt.savefig('p22_halfrange_sine.png', dpi=120)
plt.show()









```
