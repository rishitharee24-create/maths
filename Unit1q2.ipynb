```python
import sympy as sp

t = sp.symbols('t', positive=True)
v = sp.Function('v')
R, C, V = 100, 0.01, 5

ode = sp.Eq(v(t).diff(t) + v(t)/(R*C), V/(R*C))
sol = sp.dsolve(ode, v(t), ics={v(0): 0})
print("v(t) =", sp.simplify(sol.rhs))

# Numerical check at a few times
v_func = sp.lambdify(t, sol.rhs, 'numpy')
for tv in [0, 0.5, 1, 2, 5]:
    print(f"v({tv}) = {v_func(tv):.4f} V")
```
