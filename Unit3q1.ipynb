```python
import sympy as sp

t, s = sp.symbols('t s', positive=True)
R, L, C, V = 5, 1, sp.Rational(1, 5), 10

# Laplace domain impedance equation: (Ls + R + 1/(Cs)) I(s) = V/s
I_s = (V/s) / (L*s + R + 1/(C*s))
I_s = sp.simplify(I_s)
print("I(s) =", I_s)

i_t = sp.inverse_laplace_transform(I_s, s, t)
print("i(t) =", sp.simplify(i_t))





```
