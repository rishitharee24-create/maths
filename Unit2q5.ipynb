```python
import sympy as sp

t = sp.symbols('t', positive=True)
q = sp.Function('q')
R, L, C, V = 6, 1, sp.Rational(1, 25), 24  # 1/C = 25

ode = sp.Eq(L*q(t).diff(t, 2) + R*q(t).diff(t) + q(t)/C, V)
sol = sp.dsolve(ode, q(t), ics={q(0): 0, q(t).diff(t).subs(t, 0): 0})
q_sol = sp.simplify(sol.rhs)
print("q(t) =", q_sol)

i_sol = sp.simplify(sp.diff(q_sol, t))
print("i(t) =", i_sol)

# Particular solution (steady-state charge) via undetermined coefficients: q_p = constant A
A = sp.symbols('A')
qp = A
eqp = sp.Eq(qp/C, V)
A_val = sp.solve(eqp, A)[0]
print("Particular solution q_p =", A_val, " (final capacitor charge)")




```
