```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

t, n = sp.symbols('t n', positive=True)
L = sp.pi
f = sp.pi - t

a0 = (2/L) * sp.integrate(f, (t, 0, L))
an = (2/L) * sp.integrate(f * sp.cos(n*sp.pi*t/L), (t, 0, L))
an_simplified = sp.simplify(an)

print("a0 =", a0)
print("a_n =", an_simplified)

an_func = sp.lambdify(n, an_simplified, 'numpy')
N = 5
coeffs = [an_func(k) for k in range(1, N+1)]
print("First 5 a_n coefficients:", coeffs)

tvals = np.linspace(0, np.pi, 300)
v_approx = np.full_like(tvals, float(a0)/2)
for k in range(1, N+1):
    v_approx += coeffs[k-1] * np.cos(k*np.pi*tvals/np.pi)

plt.plot(tvals, np.pi - tvals, '--', label='Original v(t)=π-t')
plt.plot(tvals, v_approx, label=f'{N}-term cosine series')
plt.legend(); plt.grid(True); plt.xlabel('t'); plt.title('Half-Range Cosine Series')
plt.savefig('p23_halfrange_cosine.png', dpi=120)
plt.show()











```
