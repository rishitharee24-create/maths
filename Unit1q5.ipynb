```python
import sympy as sp

t = sp.symbols('t', positive=True)
v = sp.Function('v')
R, C = 50, 0.001
vin = 10*sp.cos(100*t)

ode = sp.Eq(R*C*v(t).diff(t) + v(t), vin)
sol = sp.dsolve(ode, v(t), ics={v(0): 0})
sol = sp.simplify(sol)
print("v(t) =", sol.rhs)

# Separate transient and steady-state parts
expanded = sp.expand(sol.rhs)
print("\nExpanded form:")
sp.pprint(expanded)

```
