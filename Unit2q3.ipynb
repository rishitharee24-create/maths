```python
import sympy as sp

t = sp.symbols('t', positive=True)
q = sp.Function('q')
R, L, C = 2, 1, sp.Rational(1, 2)

# Using charge q(t): L q'' + R q' + q/C = v(t)
ode = sp.Eq(L*q(t).diff(t, 2) + R*q(t).diff(t) + q(t)/C, 20*sp.cos(2*t))
sol = sp.dsolve(ode, q(t), ics={q(0): 0, q(t).diff(t).subs(t, 0): 0})
q_sol = sp.simplify(sol.rhs)
print("q(t) =", q_sol)

i_sol = sp.simplify(sp.diff(q_sol, t))
print("i(t) = q'(t) =", i_sol)

# Steady-state (particular) part separately
A, B = sp.symbols('A B')
qp = A*sp.cos(2*t) + B*sp.sin(2*t)
eq_particular = sp.Eq(L*qp.diff(t,2) + R*qp.diff(t) + qp/C, 20*sp.cos(2*t))
coeffs = sp.solve(sp.expand_trig(eq_particular.lhs - eq_particular.rhs), [A, B])
print("Steady-state coefficients (A,B):", coeffs)


```
