```python
import sympy as sp

t, s = sp.symbols('t s', positive=True)
R, L = 4, 2

# Laplace transform of v(t) = 8*u(t-3) is 8*exp(-3s)/s
V_s = 8*sp.exp(-3*s)/s

# (Ls + R) I(s) = V(s)  =>  I(s) = V(s) / (Ls+R)
I_s = V_s / (L*s + R)
I_s = sp.simplify(I_s)
print("I(s) =", I_s)

i_t = sp.inverse_laplace_transform(I_s, s, t)
i_t = sp.simplify(i_t)
print("i(t) =", i_t)

# Evaluate at sample times
i_func = sp.lambdify(t, i_t, 'numpy')
for tv in [0, 1, 3, 3.001, 5, 10]:
    print(f"i({tv}) = {i_func(tv):.4f} A")




```
