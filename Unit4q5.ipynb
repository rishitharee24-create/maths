```python
import sympy as sp

Va, Vb = sp.symbols('Va Vb')
R1 = R2 = R3 = R4 = 10
Rg = 5
Vs = 5

eq1 = sp.Eq((Va - Vs)/R1 + Va/R3 + (Va - Vb)/Rg, 0)
eq2 = sp.Eq((Vb - Vs)/R2 + Vb/R4 + (Vb - Va)/Rg, 0)

sol = sp.solve([eq1, eq2], [Va, Vb])
print("Solution:", sol)

I_g = (sol[Va] - sol[Vb]) / Rg
print("Galvanometer current I_g =", I_g)

# Balance check
print("\nBalance ratio R1/R3 =", R1/R3, " R2/R4 =", R2/R4)
if sp.simplify(sol[Va] - sol[Vb]) == 0:
    print("Bridge is BALANCED -> Va = Vb -> unique solution still exists,")
    print("but galvanometer branch carries zero current (decoupled).")
else:
    print("Bridge is UNBALANCED -> Va != Vb, unique non-trivial solution.")









```
