```python
import numpy as np
import matplotlib.pyplot as plt

A_amp, R, C = 10, 1, 1
N_harmonics = 7
t = np.linspace(0, 2*np.pi, 1000)

v_in = np.zeros_like(t)
v_out = np.zeros_like(t)

for n in range(1, N_harmonics*2, 2):  # odd harmonics 1,3,5,...
    bn = 4*A_amp/(np.pi*n)
    v_in += bn * np.sin(n*t)

    # RC low-pass response to sin(n*omega*t): H(jw) = 1/(1+jwRC)
    omega = n
    H_mag = 1/np.sqrt(1+(omega*R*C)**2)
    phase = -np.arctan(omega*R*C)
    v_out += bn * H_mag * np.sin(n*t + phase)

plt.plot(t, v_in, label='Input (square wave approx)')
plt.plot(t, v_out, label='Output v_C(t)')
plt.legend(); plt.grid(True); plt.xlabel('t'); plt.title('Square Wave Through RC Filter')
plt.savefig('p21_fourier_rc.png', dpi=120)
plt.show()









```
