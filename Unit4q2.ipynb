```python
import numpy as np

A = np.array([
    [10, -4, -2],
    [-4, 12, -6],
    [-2, -6, 15]
], dtype=float)
b = np.array([12, 0, 0], dtype=float)

det_A = np.linalg.det(A)
print(f"det(A) = {det_A:.4f}")

if abs(det_A) > 1e-10:
    print("Matrix non-singular -> unique solution exists")
    I = np.linalg.solve(A, b)
    print("Mesh currents I1, I2, I3 =", I)
else:
    print("Matrix singular -> no unique solution (check for infinite/no solutions)")







```
