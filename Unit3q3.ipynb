```python
import sympy as sp

t, s = sp.symbols('t s', positive=True)
I1, I2 = sp.symbols('I1 I2')

# Laplace-domain algebraic system
eq1 = sp.Eq((2*s + 3)*I1 - I2, 10/s)
eq2 = sp.Eq(-I1 + (s + 2)*I2, 0)

sol = sp.solve([eq1, eq2], [I1, I2])
I1_s = sp.simplify(sol[I1])
I2_s = sp.simplify(sol[I2])
print("I1(s) =", I1_s)
print("I2(s) =", I2_s)

i1_t = sp.simplify(sp.inverse_laplace_transform(I1_s, s, t))
i2_t = sp.simplify(sp.inverse_laplace_transform(I2_s, s, t))
print("\ni1(t) =", i1_t)
print("i2(t) =", i2_t)



```
