```python
import sympy as sp

x = sp.symbols('x', positive=True)
y = sp.Function('y')

ode = sp.Eq(x**2 * y(x).diff(x, 2) + 3*x*y(x).diff(x) + y(x), 0)
sol = sp.dsolve(ode, y(x), ics={y(1): 2, y(x).diff(x).subs(x, 1): -1})
print("y(x) =", sp.simplify(sol.rhs))

# General-solution check via trial y = x^m
m = sp.symbols('m')
char_eq = m*(m-1) + 3*m + 1   # m^2 + 2m + 1 = 0
roots = sp.solve(char_eq, m)
print("Euler-Cauchy auxiliary roots:", roots)  # repeated root -> y = x^m (C1 + C2 ln x)

```
