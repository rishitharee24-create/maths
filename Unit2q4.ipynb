```python
import numpy as np
import matplotlib.pyplot as plt

R, L, C, V0 = 5, 0.5, 200e-6, 10

omega0 = 1/np.sqrt(L*C)
print(f"Resonant frequency omega0 = {omega0:.4f} rad/s")
print(f"Resonant frequency f0 = {omega0/(2*np.pi):.4f} Hz")

omega = np.linspace(1, 500, 2000)
Z = np.sqrt(R**2 + (omega*L - 1/(omega*C))**2)
I_amp = V0 / Z

plt.plot(omega, I_amp)
plt.axvline(omega0, color='r', linestyle='--', label=f'omega0 = {omega0:.1f}')
plt.xlabel('omega (rad/s)'); plt.ylabel('|I(omega)| (A)')
plt.title('RLC Resonance Curve'); plt.legend(); plt.grid(True)
plt.savefig('p9_resonance.png', dpi=120)
plt.show()

print(f"Max current at resonance = {V0/R:.4f} A")


```
