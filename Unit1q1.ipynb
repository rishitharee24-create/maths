```python
import numpy as np
import matplotlib.pyplot as plt

R, L, V = 10, 2, 12
h, t_end = 0.05, 1.0
n = int(t_end / h)

# Euler's method
t = np.zeros(n+1)
i_euler = np.zeros(n+1)
for k in range(n):
    di_dt = (V - R * i_euler[k]) / L
    i_euler[k+1] = i_euler[k] + h * di_dt
    t[k+1] = t[k] + h

# Exact analytical solution: i(t) = (V/R)(1 - e^(-Rt/L))
i_exact = (V/R) * (1 - np.exp(-R*t/L))

for k in range(0, n+1, 4):
    print(f"t={t[k]:.2f}  Euler={i_euler[k]:.4f}  Exact={i_exact[k]:.4f}")

plt.plot(t, i_euler, 'o-', label='Euler')
plt.plot(t, i_exact, '--', label='Exact')
plt.xlabel('t (s)'); plt.ylabel('i(t) (A)'); plt.legend(); plt.title('RL Circuit - Euler vs Exact')
plt.grid(True); plt.savefig('p1_rl_euler.png', dpi=120)
plt.show()
```
