```python
import sympy as sp

t, s = sp.symbols('t s', positive=True)
R, C = 10, sp.Rational(1, 10)

# Laplace of 5*delta(t) = 5
V_in_s = 5

# (RCs + 1) V(s) = 5  => V(s) = 5/(RCs+1)
V_s = V_in_s / (R*C*s + 1)
V_s = sp.simplify(V_s)
print("V(s) =", V_s)

v_t = sp.simplify(sp.inverse_laplace_transform(V_s, s, t))
print("v(t) =", v_t)

# Current i(t) = C dv/dt
i_t = sp.simplify(C*sp.diff(v_t, t))
print("i(t) =", i_t)

# Integration property: Q(s) = I(s)/s  -> q(t) = integral of i(t)
I_s = sp.laplace_transform(i_t, t, s, noconds=True)
Q_s = sp.simplify(I_s / s)
q_t = sp.simplify(sp.inverse_laplace_transform(Q_s, s, t))
print("q(t) (via integration property) =", q_t)
print("q(t) (direct integration)        =", sp.simplify(sp.integrate(i_t, (t, 0, t))))




```
