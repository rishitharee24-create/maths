```python
import numpy as np

A = np.array([
    [1/10+1/20+1/30, -1/20, -1/30],
    [-1/20, 1/20+1/40, -1/40],
    [-1/30, -1/40, 1/30+1/40+1/50]
])
b = np.array([2, 0, 1], dtype=float)

def gauss_elimination(A, b):
    A = A.astype(float).copy()
    b = b.astype(float).copy()
    n = len(b)
    # Forward elimination
    for k in range(n-1):
        for i in range(k+1, n):
            factor = A[i, k] / A[k, k]
            A[i, k:] -= factor * A[k, k:]
            b[i] -= factor * b[k]
    # Back substitution
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = (b[i] - A[i, i+1:] @ x[i+1:]) / A[i, i]
    return x

V = gauss_elimination(A, b)
print("Node voltages V1, V2, V3 =", V)
print("Check with numpy.linalg.solve:", np.linalg.solve(A, b))





```
