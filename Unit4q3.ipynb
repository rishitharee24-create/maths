```python
import numpy as np

A = np.array([
    [2, -0.5],
    [-0.5, 2]
])

eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues (omega^2):", eigvals)

omegas = np.sqrt(eigvals)
print("Natural frequencies omega1, omega2 (rad/s):", omegas)

print("\nEigenvectors (mode shapes):")
for i in range(len(eigvals)):
    print(f"  Mode {i+1}: omega={omegas[i]:.4f}, eigenvector={eigvecs[:, i]}")








```
