```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

t = sp.symbols('t', positive=True)
i = sp.Function('i')
R, L = 5, 1

ode = sp.Eq(L*i(t).diff(t) + R*i(t), 0)
sol = sp.dsolve(ode, i(t), ics={i(0): 2})
print("i(t) =", sol.rhs)

i_func = sp.lambdify(t, sol.rhs, 'numpy')
tvals = np.linspace(0, 2, 100)
plt.plot(tvals, i_func(tvals))
plt.xlabel('t (s)'); plt.ylabel('i(t) (A)'); plt.title('RL Discharge')
plt.grid(True); plt.savefig('p4_rl_discharge.png', dpi=120)
plt.show()

```
